```python
from main import *
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
```

When rolling 6 dice 10^4 times we do see that they are in fact fair

```python
data = Counter()
for _ in range(10**4):
    data += Counter(roll_dice(6))

sns.barplot(x=list(data.keys()), y=list(data.values()), hue=list(data.keys()), palette="bright", legend=False)
plt.xlabel("Dice Value")
plt.ylabel("Occurences")
plt.title("10^4 Rolls of 6 Dice")
plt.show()
```

A more intresting aspect of farkel to investigate is survival-chance. During a round the player may choose to roll again with fewer dice, which could either result in more points or getting farkeled (losing all obtained points in the round due to not rolling any scoring dice). The chance to bust at the beginning of the turn is incredibly low, and even on 3 dice you still have a very good chance at surviving.

```python
data = {}
for i in range(1, 7):
    data[i] = 0
    for _ in range(10**3):
        data[i] += is_valid(roll_dice(i))
data = {k: v/10**3 for k,v in data.items()}

sns.barplot(x=list(data.keys()), y=list(data.values()), hue=list(data.keys()), palette="bright", legend=False)
plt.xlabel("Number of Dice")
plt.ylabel("Survival Chance")
plt.title("10^3 Rolls of N Dice")
plt.show()
```
